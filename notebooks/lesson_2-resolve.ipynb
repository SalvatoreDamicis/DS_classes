#### Analisi delle durate dei film
Quanto dura un film, e quanto dura un movies? 

```python
df_show = df_netflix[df_netflix['type']=='TV Show'].copy()
df_show['duration_int'] = df_show['duration'].apply(lambda x: int(x.split(' ')[0]))
sns.histplot(df_show['duration_int'])
```

```python
df_movie = df_netflix[df_netflix['type']=='Movie'].copy()
df_movie['duration_int'] = df_movie['duration'].str.replace(' min', '').astype(int)
sns.histplot(df_movie['duration_int'])
```
